```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
irisdata=load_iris()
print("Shape of the data:{}".format(irisdata['data'].shape))
```

```python
print("Type of target:{}".format(type(irisdata['target'])))
print("Shape of target:{}".format(irisdata['target'].shape))
print("Target:{}".format(irisdata['target']))
```

```python
X_train,X_test,y_train,y_test=train_test_split(irisdata['data'],irisdata['target'],random_state=0)
print("Shape of x_train:{}".format(X_train.shape))
```

```python
print("Shape of y_train:{}".format(y_train.shape))
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
x_new=np.array([[6,2,2,0.4]])
pred= knn.predict(x_new)
print("Target is:{}".format(pred))
print("Target name is:{}".format(irisdata['target_names'][pred]))
```

```python

y=knn.predict(X_test)
print("Test set predictions:{}".format(y))
print("Accuracy:{:.2f}".format(np.mean(y==y_test)))
```
